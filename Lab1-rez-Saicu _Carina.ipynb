## Exercițiul 1

Fie semnalele continue $x(t) = \cos (520 \pi t + \pi/3)$, $y(t) = \cos (280 \pi t - \pi/3)$ și $z(t) = \cos (120 \pi t + \pi/3)$.

- În Python, simulați axa reală de timp printr-un șir de numere suficient de apropiate, spre exemplu $[0:0.0005:0.03]$ (adică intervalul $[0, 0.03]$, discretizat cu un pas de $0.0005$).

```python
t = np.arange(0,0.03,0.0005)

plt.plot(t)
plt.xlabel('Timp')
plt.ylabel('Valoare')
plt.title('Axa reala de timp')
plt.show()
```

- Construiți semnalele $x(t)$, $y(t)$ și $z(t)$ și afișați-le grafic, în câte un subplot.

```python
x = np.cos(520 * np.pi * t + np.pi/3)

plt.subplot(131) # 1 - rand, 3 - coloane, 1 - primul subplot
plt.plot(t, x)
plt.title('x')

y = np.cos(280 * np.pi * t - np.pi/3)

plt.subplot(132)
plt.plot(t, y)
plt.title('y')

z = np.cos(120 * np.pi * t + np.pi/3)

plt.subplot(133)
plt.plot(t, z)
plt.title('z')

plt.tight_layout() # sa nu se suprapuna

plt.show()
```

- Eșantionați semnalele cu o frecvență de 200 Hz pentru a obține $x[n]$, $y[n]$ și $z[n]$ și afișați-le grafic, în câte un subplot.

```python
#inainte

# nr de esantioane = durata / pas = 0.03 / 0.0005 = 60
# frecventa = 60 / 0.03 = 2000 Hz

#acum 

# avem nevoie de frecventa = 200 Hz  => pas = 0.005

s = np.arange(0,0.03,0.005)

x = np.cos(520 * np.pi * s + np.pi/3)

plt.subplot(131) # 1 - rand, 3 - coloane, 1 - primul subplot
plt.plot(s, x)
plt.title('x')

y = np.cos(280 * np.pi * s - np.pi/3)

plt.subplot(132)
plt.plot(s, y)
plt.title('y')

z = np.cos(120 * np.pi * s + np.pi/3)

plt.subplot(133)
plt.plot(s, z)
plt.title('z')

plt.tight_layout() # sa nu se suprapuna

plt.show()
```

## Exercițiul 2

Generați semnale descrise mai jos și afișați-le grafic, fiecare într-un plot. Nu aveți voie să folosiți funcțiile predefinite din `scipy.signal`.

- Un semnal **sinusoidal** de frecvență 400 Hz, care să conțină 1600 de eșantioane.

```python
# frecventa = nr de esantioane / durata => 400 Hz = 1600 / durata  => durata = 1600 / 400 => durata = 4 s
# nr esantioane = durata / pas => 1600 = 4 / pas => pas = 4 / 1600 => pas = 0.0025

t = np.arange(0,4,0.0025)
semnal_sinusoidal = np.sin(2 * np.pi * 400 * t)

plt.plot(t, semnal_sinusoidal)
plt.title('Semnal Sinusoidal de 400 Hz')
plt.show()
```

- Un semnal **sinusoidal** de frecvență 800 Hz, care să dureze 3 secunde.

```python
# frecventa = nr de esantioane / durata => 800 Hz = nr de esantioane / 3  => nr de esantioane = 800 * 3  => nr de esantioane = 2400
# nr esantioane = durata / pas => 2400 = 3 / pas => pas = 3 / 2400 => pas = 0.00125


t = np.arange(0,3,0.00125)
semnal_sinusoidal = np.sin(2 * np.pi * 800 * t)

plt.plot(t, semnal_sinusoidal)
plt.title('Semnal Sinusoidal de 800 Hz')
plt.show()
```

- Un semnal de tip **sawtooth** de frecvență 240 Hz (puteți folosi funcțiile `numpy.floor` sau `numpy.mod`).

```python
# frecventa = nr de esantioane / durata
# vom alege durata de 2 secunda
#  240 = nr de esantioane / 2 => nr de esantioane = 480
# nr esantioane = durata / pas => 480 = 2 / pas => pas = 2 / 480 => pas = 0.00416

t = np.linspace(0,2,240)
semnal_sawtooth = (np.mod(t * 240, 1) - 0.5) * 2.0

plt.plot(t, semnal_sawtooth)
plt.title('Semnal "Sawtooth" de 240 Hz')
plt.show()
```

- Un semnal de tip **square** de frecvență 300 Hz (puteți folosi funcția `numpy.sign`).

```python
# nr de esantioane  = durata * frecventa => daca alegem durata = 2 s => nr de esantioane = 2 * 300 Hz => nr de esantioane = 600 Hz

t = np.linspace(0, 2, 600)
semnal_square = np.sign(np.sin(2 * np.pi * 300 * t))

plt.plot(t, semnal_square)
plt.title('Semnal "Square" de de 300 Hz')
plt.ylim(-1.5, 1.5)  # setati limitele pentru a afisa corect semnalul patrat
plt.show()
```

- Un semnal 2D **aleator**. Creați un `numpy.array` de dimensiune 128x128 și inițializați-l aleator, folosind `numpy.random.rand(x, y)`, unde `x` și `y` reprezintă numărul de linii, respectiv de coloane.

- Un semnal 2D **la alegerea voastră**. Creați un `numpy.array` de dimensiune 128x128 și inițializați-l folosind o procedură creată de voi.
Utilizați, spre exemplu, funcțiile `numpy.zeros()` și `numpy.ones()`.

```python
x = 128
y = 128

semnal_2d = np.zeros((x, y))

# in aceasta configuratie, fiecare linie alternativa va fi neagra
semnal_2d[1::2, :] = 1  # setam linile pare la 1 (alb)
semnal_2d[0::2, :] = 0  # setam linile impare la 0 (negru)

plt.imshow(semnal_2d, cmap='gray', interpolation='nearest')
plt.title('Model cu dungi orizontale')
plt.colorbar()
plt.show()
```

## Exercițiul 3

Un semnal este înregistrat cu o frecvență de eșantionare de 2000 Hz.

- Care este intervalul de timp între două eșantioane?

- Dacă un eșantion este memorat pe 4 biți, câți bytes vor ocupa 1 oră de achiziție?

```python
# date enunt

frec_es = 2000
durata_ore = 1
biti_esantion = 4

# rezolvare

interval_timp = 1 / frec_es #intervalul de timp intre doua esantioane
numar_es_ora = durata_ore * 3600 * frec_es # 60 * 60 = 3600
total_biti = numar_es_ora * biti_esantion
total_bytes = total_biti / 8 # 1 byte = 8 biti

print(f"Intervalul de timp intre esantioane: {interval_timp} secunde")
print(f"Numarul de bytes in 1 ora de achizitie: {total_bytes} bytes")
```
